## 3. bad 集中度


- **$P(Bad|Cat)$（類別純度）**：就像 Precision。當你指出這個類別有問題時，裡面真的都是壞片的機率。
- **$P(Cat|Bad)$（壞片集中度）**：就像 Recall。所有的壞片中，被這個類別抓到了多少。
- F1-score 是 Precision 和 Recall 的調和平均數 (Harmonic Mean)

- 幾何平均比較寬容：如果一個類別純度 1.0，但集中度只有 0.1（代表這是一個很小但很純的異常），幾何平均會給出 $\sqrt{0.1} \approx 0.31$。
- F1-score 比較嚴格：同樣情況下，F1 會給出 $2 \times \frac{0.1}{1.1} \approx 0.18$。它會強制要求兩個指標都要達到一定水準，總分才會高。
